```python
# Importing necessary libraries

# data manipulation
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv('../Rossmann_Store_Sales/Data/results.csv')
```

```python
results
```

```python
results.sort_values('RMSPE on Kaggle', inplace=True, ascending=False)
```

```python
results.reset_index(drop=True, inplace=True)
```

```python
results
```

```python
plt.figure(figsize=(15,8))

plot = sns.lineplot(x=results.Model,y=results['RMSPE on Kaggle'], label='RMSPE', marker="o", sort=False, **{"color": '#247AB2', "markersize": 10, "linestyle": "--"})

plt.xlabel("Алгоритам")
plt.ylabel('RMSPE')

plt.setp(plot.get_xticklabels(), rotation=70)

plt.legend()
```
